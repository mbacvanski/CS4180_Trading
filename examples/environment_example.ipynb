```python
import matplotlib.pyplot as plt
from easydict import EasyDict

from environment import StocksEnv
```

```python
env = StocksEnv(EasyDict({
    "env_id": 'stocks-v0', "eps_length": 30,
    "window_size": 20, "train_range": None, "test_range": None,
    "stocks_data_filename": 'STOCKS_GOOGL'
}))
env.seed(0)
```

```python
from environment import Actions

state = env.reset()
for i in range(29):
    print(f'Time step {i} ===================================================')
    print(f'State: {state}')
    action = env.random_action()
    print(f'Action: {Actions(action).name}')
    step = env.step(action)

    print(f'Next state: {step.obs}')
    print(f'Reward: {step.reward}')
```
